```python
from PIL import Image
from IPython.display import display
import cv2
```

```python
#Reading the image
image=cv2.imread("GokuSaiyan.jpg")
display(Image.fromarray(image))
```

The above image is having bluish shade because the opencv reads the image in BGR format

```python
#Displaying the original image with resizing 
image=cv2.imread("GokuSaiyan.jpg")
Original_Image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
Resized_Original_Image=cv2.resize(Original_Image,(600,550))
display(Image.fromarray(Resized_Original_Image))
```

```python
#Converting Original image to Grey scale image
Grey_Goku_Saiyan=cv2.cvtColor(Resized_Original_Image,cv2.COLOR_RGB2GRAY)
display(Image.fromarray(Grey_Goku_Saiyan))
```

```python
#Inverting the new grayscale image of Goku_Saiyan
inverted_Goku_Saiyan = 255 - Grey_Goku_Saiyan
display(Image.fromarray(inverted_Goku_Saiyan))
```

```python
#Smoothing the image using Gaussian Filtering to remove noise from the image 
Smoothed_Goku_Saiyan=cv2.GaussianBlur(inverted_Goku_Saiyan,(11,11),0)
display(Image.fromarray(Smoothed_Goku_Saiyan))
#Displaying the blurred image due to the removal of noise
```

```python
#Inverting the Smoothed image of Goku_Saiyan and converting into pencil sketch
inverted_Smoothed_Goku_Saiyan=255-Smoothed_Goku_Saiyan
PencilSketch_Of_Goku_Saiyan=cv2.divide(Grey_Goku_Saiyan,inverted_Smoothed_Goku_Saiyan,scale=256.0)
display(Image.fromarray(PencilSketch_Of_Goku_Saiyan))
```

```python
display(Image.fromarray(PencilSketch_Of_Goku_Saiyan))
display(Image.fromarray(Resized_Original_Image))
```
